```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pylab
import matplotlib.patches
import matplotlib.lines
import matplotlib.path

results_up = {"Angle-up": [], "X-coord-up": [], "Y-coord-up": []}
results_down = {"Angle-down": [], "X-coord-down": [], "Y-coord-down": []}
results_up["Angle-up"] = np.array([-30, 0, 30, 45, 60, 90])
results_down["Angle-down"] = np.array([60, 45, 30, 0, -30])
results_up["X-coord-up"] = np.array([30, 29, 28, 26, 26, 26])
results_up["Y-coord-up"] = np.array([35, 41, 45, 47, 50, 50])
results_down["X-coord-down"] = np.array([25, 25, 25, 26, 29])
results_down["Y-coord-down"] = np.array([50, 50, 48, 43, 36])
#===========================================================================
pd_results = pd.DataFrame(results_up)
pd_results = pd.DataFrame(results_down)

#print(pd_results)
plt.figure(figsize=(10,10))

plt.plot(results_up["X-coord-up"], results_up["Y-coord-up"], '^', color='red')
for i in range (len(results_up["Angle-up"])):
    plt.annotate(results_up["Angle-up"][i], xy=(results_up["X-coord-up"][i],results_up["Y-coord-up"][i]), xytext=(results_up["X-coord-up"][i]+0.2,results_up["Y-coord-up"][i]+0.2),fontsize=12)
    
plt.plot(results_down["X-coord-down"], results_down["Y-coord-down"], 'v', color='blue')
for i in range (len(results_down["Angle-down"])):
    plt.annotate(results_down["Angle-down"][i], xy=(results_down["X-coord-down"][i],results_down["Y-coord-down"][i]), xytext=(results_down["X-coord-down"][i]-0.2,results_down["Y-coord-down"][i]-0.2),fontsize=12)
#plt.plot(CN40, HT40, linestyle='-', marker='o', label = u'A = 40', color='r')
#plt.plot(CN36, HT36, linestyle='-', marker='o', label = u'A = 36', color='b')
plt.title('Test_2')

#plt.yscale('log')
plt.grid(True)
#plt.legend(loc='upper left')
gridsize=(10,10)

plt.xlabel('X-difference, mm')
plt.ylabel('Y-difference, mm')
#save('pic_12_1_1', fmt='png')
plt.savefig('test_2', fmt='pdf')

#plt.hlines(u, l, r, linestyle='-', colors='b')
#plt.hlines(d, l, r, linestyle='-', colors='b')
#plt.vlines(l, d, u, linestyle='-', colors='b')
#plt.vlines(r, d, u, linestyle='-', colors='b')

plt.show()
```
